```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=4)
km
```

```python
y_predicted=km.fit_predict(df[['Age','Spending Score (1-100)']])
y_predicted
```

```python
df['cluster1']=y_predicted
df.head()
```

```python
df1=df[df.cluster1==0]
df2=df[df.cluster1==1]
df3=df[df.cluster1==2]
df4=df[df.cluster1==3]

plt.scatter(df1.Age,df1['Spending Score (1-100)'],color='red')
plt.scatter(df2.Age,df2['Spending Score (1-100)'],color='green')
plt.scatter(df3.Age,df3['Spending Score (1-100)'],color='blue')
plt.scatter(df4.Age,df4['Spending Score (1-100)'],color='black')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')

plt.xlabel('Age')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```

```python
km.cluster_centers_
```

```python
k_rng=range(1,10)
sse=[]
for x in k_rng:
    km=KMeans(n_clusters=x)
    km.fit(df[['Age','Spending Score (1-100)']])
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.xlabel('K')
plt.ylabel('Sum of squared errors')
plt.plot(k_rng,sse)
plt.show()
```

```python
#FROM THE ELBOW GRAPH, WE CAN GROUP THE DATA INTO 4 SUITABLE CLUSTERS
#FROM AGE VS SPENDING SCORE(1-100) GRAPH, WE INFER THAT
#  - SPENDING SCORE IS HIGH WITHIN THE AGE LIMIT (20-40)
#  - THE MAXIMUM SPENDING SCORE IS UPTO 60 FOR AGE 50-70
#  - THE AGE LIMIT 30-40 HAS THE MOST FREQUENCY OF SPENDING SCORE
```

```python
dataframe=pd.read_csv("Mall_Customers.csv")
dataframe.describe()
```

```python
dataframe['Gender'].replace(['Male','Female'],[0,1],inplace=True)
```

```python
dataframe.head()
```

```python
from sklearn.cluster import KMeans
km2=KMeans(n_clusters=4)
km2
```

```python
y_predicted2=km2.fit_predict(dataframe[['Age','Gender']])
y_predicted2
```

```python
dataframe['cluster']=y_predicted2
dataframe.head()
```

```python
dataframe1=dataframe[dataframe.cluster==0]
dataframe2=dataframe[dataframe.cluster==1]
dataframe3=dataframe[dataframe.cluster==2]
dataframe4=dataframe[dataframe.cluster==3]

plt.scatter(dataframe1.Age,dataframe1['Gender'],color='red')
plt.scatter(dataframe2.Age,dataframe2['Gender'],color='green')
plt.scatter(dataframe3.Age,dataframe3['Gender'],color='blue')
plt.scatter(dataframe4.Age,dataframe4['Gender'],color='black')
plt.scatter(km2.cluster_centers_[:,0],km2.cluster_centers_[:,1],color='purple',marker='*',label='centroid')

plt.xlabel('Age')
plt.ylabel('Gender')
plt.legend()
plt.show()
```

```python
km2.cluster_centers_
```

```python
k_rng=range(1,10)
sse=[]
for x in k_rng:
    km2=KMeans(n_clusters=x)
    km2.fit(dataframe[['Age','Gender']])
    sse.append(km2.inertia_)
```

```python
sse
```

```python
plt.xlabel('K')
plt.ylabel('Sum of squared errors')
plt.plot(k_rng,sse)
plt.show()
```
